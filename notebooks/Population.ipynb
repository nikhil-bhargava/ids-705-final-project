```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('/Users/caleboneel/Downloads/archiveBnb/train.csv')
city = pd.read_csv('./worldcities.csv')
city = city.rename(columns={"city": "city1"})
city.columns
```

```python
city = city[['city1', 'population', 'id']]
df['city'] = df['city'].replace({'NYC': 'New York', 'LA' : 'Los Angeles', 'SF' : 'San Francisco', 'DC' : "Washington"})
```

```python
ID = {
    'New York': 1840034016, 
    'San Francisco': 1840021543,
    'Los Angeles': 1840020491,
    'Washington': 1840006060, 
    'Boston': 1840000455,
    'Chicago' : 1840000494
}
```

```python
df['City_ID'] = df['city'].replace(ID)
df
```

```python
df = df.merge(city, how = 'left', left_on = 'City_ID', right_on = 'id')
```

```python
pop = df[['city', 'population']]
pop.population.unique()
```

```python
df['highPop'] = [1 if x >=10000000 else 0 for x in df['population']]
df['medPop'] = [1 if (x <10000000 and x >=5000000)  else 0 for x in df['population']]
df['lowPop'] = [1 if x < 5000000 else 0 for x in df['population']]
```
