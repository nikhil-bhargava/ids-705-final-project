```python
import pandas as pd
import numpy as np
from collections import Counter
from collections import defaultdict
```

```python
df = pd.read_csv("../data/raw/train.csv")

df.head()
```

```python
df.shape
```

```python
amenities = df["amenities"]
```

```python
def handle_entry(s):
    # remove curly brace
    s = s.strip('{').strip('}')
    # split
    s_list = s.split(",")
    # remove extra quote
    for i, curr_s in enumerate(s_list):
        s_list[i] = s_list[i].strip('"')
    return s_list
```

```python
all_amenities = list()
```

```python
for i in range(amenities.shape[0]):
    s = str(amenities[i])
    new_list = handle_entry(s)
    new_list = [ss for ss in new_list if not ss.startswith('translation missing:')]
    all_amenities += new_list
```

```python
unique_amenity_cnt = Counter(all_amenities)
```

```python
amentity_pct = defaultdict(list)
```

```python
for key in unique_amenity_cnt:
    pct = round(unique_amenity_cnt[key]/df.shape[0], 2)
    amentity_pct[key].append(unique_amenity_cnt[key])
    amentity_pct[key].append(pct)
```

```python
unique_list = sorted(list(amentity_pct.keys()))
```
